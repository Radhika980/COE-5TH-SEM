```python
p1 = np.array([[1,2],[3,4]])
print("Array p1:\n",p1)
q1 = np.array([[4,5],[6,7]])
print("Array q1:\n",q1)
print("Product of p1 and q1:\n",np.matmul(p1,q1))

output = []
for a,b in zip(p1,q1):
    cov = np.cov(a, b)
    output.append(cov[0][1])
print("Covariance between p1 and q1:\n",output)


p2 = np.array([[1,2],[2,3],[4,5]])
print("\nArray p2:\n",p2)
q2 = np.array([[4,5,1],[6,7,2]])
print("Array q2:\n",q2)
print("Product of p2 and q2:\n",np.matmul(p2,q2))
q2 = np.array([[4,5,1],[6,7,2]]).reshape(3,2)
output2 = []
for a,b in zip(p2,q2):
    cov = np.cov(a, b)
    output2.append(cov[0][1])
print("Covariance between p2 and q2:\n",output2)
```
